## Step 7: Vectorization

Next, we will vectorize the tweet content by generating a document-term matrix with TF-IDF scoress.
- We remove English stopwords.
- max_df=0.8
- min_df=0.02

````python
from sklearn.feature_extraction.text import TfidfVectorizer

# 💡 Hint: See class notebook Section 3.3 "Vectorization" – the TfidfVectorizer block
#         Here: stop_words='english', max_df=0.8, min_df=0.02
#         Run on tweets_openAI first; repeat for tweets_nvidia in Steps 9 & 10

docs = tweets_openAI['text']

tfidf_vectorizer = TfidfVectorizer(
    stop_words='english',
    max_df=0.8,
    min_df=0.02
)

tfidf = tfidf_vectorizer.fit_transform(docs)

tfidf_df = pd.DataFrame.sparse.from_spmatrix(
    tfidf,
    columns=tfidf_vectorizer.get_feature_names_out()
)

tfidf_df
````

## Step 8. Topic Discovery
We use non-negative matrix factorization to generate 20 topics.
For each topic, we use the top 3 words to describe the topic.

````python
# Apply NMF
from sklearn.decomposition import NMF

# 💡 Hint: See class notebook Section 3.4 "Topic Modeling" – use n_components=20 here (class used 7)
#         The for-loop to print topic names is also shown there

from sklearn.decomposition import NMF

# create model
nmf_model = NMF(n_components=20, random_state=0)

# fit model
nmf_model.fit(tfidf)

# document-topic matrix
W = nmf_model.fit_transform(tfidf)

# display topics
feature_names = tfidf_vectorizer.get_feature_names_out()
topic_names = []

for topic_index, topic in enumerate(nmf_model.components_):
    top_words = pd.Series(topic, index=feature_names).nlargest(3).index.tolist()
    top_words_string = " ".join(top_words)
    print("Topic #" + str(topic_index) + ": " + top_words_string)
    topic_names.append(top_words_string)

# create topic dataframe
topic_df = pd.DataFrame(W, columns=topic_names)
topic_df
````

## Step 9 Linear Regression
Model
Construct out outcome variable $log(Likes)=log(Likes+1)$, run the model, and evaluate the model.


````python
# Code for estimating the linear regression model with the outcome being log(Likes)
# 💡 Hint: See class notebook Section 3.5 "Linear Regression"
#         Outcome: log(public_metrics.like_count + 1)  →  np.log1p(...)
#         Predictors: topic columns from W matrix (use pd.DataFrame(W, columns=topic_names))
#         Use pingouin: pg.linear_regression(X, y)

import pingouin as pg

# Step 1: Create model dataframe (topics only)
df_model = topic_df.copy()

# Step 2: Add outcome variable = log(likes + 1)
df_model['log_likes'] = np.log1p(tweets_openAI['public_metrics.like_count'])

# Step 3: Run linear regression
result = pg.linear_regression(
    df_model.drop(columns='log_likes'),   # X = topic columns
    df_model['log_likes']                # Y = outcome
)
````

````python
# Code for results production
# 💡 Hint: See class notebook Section 3.5 – print the regression results table (.round(3))

result_display = result[['names', 'coef', 'pval']].round(3)

print("Regression Results: Which Topics Drive Likes?\n")
print(result_display)

print("\n How to read this:")
print("- Positive 'coef' = this topic increases likes")
print("- Negative 'coef' = this topic decreases likes")
print("- 'pval' < 0.05 = statistically significant")
````
